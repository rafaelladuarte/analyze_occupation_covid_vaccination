```python
import pandas as pd
```

```python
path = 'dados/owid-covid-data.csv'
```

```python
df = pd.read_csv(path)
df
```

```python
df.info()
```

```python
df['location'].unique()
```

```python
df = df[df['location'] == 'Brazil']
df
```

```python
df_vac = df[['date','total_cases','new_cases','total_deaths','new_deaths','total_vaccinations','people_vaccinated','new_vaccinations']]
df_vac
```

```python
df_vac.fillna("0", inplace=True)
```

```python
df_vac
```

```python
df_vac["datacode"] = df_vac['date'].apply(lambda x : x.replace('-',''))
df_vac
```

```python
df_vac.info()
```

```python
df_vac['total_cases'] = df_vac['total_cases'].astype('int')
df_vac['new_cases'] =df_vac['new_cases'].astype('int')
df_vac['total_deaths'] = df_vac['total_deaths'].astype('int')
df_vac['total_vaccinations'] = df_vac['total_vaccinations'].astype('int')
df_vac['people_vaccinated'] = df_vac['people_vaccinated'].astype('int')
df_vac['new_vaccinations'] = df_vac['new_vaccinations'].astype('int')
df_vac['new_deaths'] = df_vac['new_deaths'].astype('int')
```

```python
df_vac.info()
```

```python
df_vac.to_csv('brasil_vacinacao_daily.csv')
```
