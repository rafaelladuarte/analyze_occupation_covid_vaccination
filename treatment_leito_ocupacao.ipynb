```python
import pandas as pd
```

```python
path2020 ='dados/LeitoOcupacao_2020.csv'
path2021 ='dados/LeitoOcupacao_2021.csv'
path2022 ='dados/LeitoOcupacao_2022.csv'
```

```python
df2020 = pd.read_csv(path2020)
df2021 = pd.read_csv(path2021)
df2022 = pd.read_csv(path2022)
```

```python
df_all = pd.concat([df2020,df2021,df2022])
```

```python
df_all.drop(
    [
      'Unnamed: 0',
      '_id',
      '_p_usuario',
      'municipioNotificacao',
      'estadoNotificacao',
      'origem',
      'excluido',
      'validado',
      '_created_at',
      '_updated_at',
     'estado',
     'municipio'

    ], 
    axis=1, inplace=True
)
df_all
```

```python
df_all['anoNotificacao'] = pd.DatetimeIndex(df_all['dataNotificacao']).year
df_all['mesNotificacao'] = pd.DatetimeIndex(df_all['dataNotificacao']).month
df_all['dayNotificacao'] = pd.DatetimeIndex(df_all['dataNotificacao']).day
```

```python
df_all
```

```python
df_all['anoNotificacao'] = df_all['anoNotificacao'].astype('str')
df_all['mesNotificacao'] = df_all['mesNotificacao'].astype('str')
df_all['dayNotificacao'] = df_all['dayNotificacao'].astype('str')
df_all.info()
```

```python
df_all = df_all.sort_values(by=['anoNotificacao','mesNotificacao','dayNotificacao'])
```

```python
df_all.fillna("0", inplace=True)
df_all
```

```python
df_all['anoNotificacao'] = df_all['anoNotificacao'].apply((lambda x: x.rjust(2, "0")))
df_all['mesNotificacao'] = df_all['mesNotificacao'].apply((lambda x: x.rjust(2, "0"))) 
df_all['dayNotificacao'] = df_all['dayNotificacao'].apply((lambda x: x.rjust(2, "0")))
df_all
```

```python
df_all['datacode'] = df_all['anoNotificacao'] +  df_all['mesNotificacao'] + df_all['dayNotificacao']
```

```python
df_all['saidaConfirmadaAltas'] = df_all['saidaConfirmadaAltas'].astype('int')
df_all['ocupacaoSuspeitoUti'] = df_all['ocupacaoSuspeitoUti'].astype('int')
df_all['ocupacaoConfirmadoCli'] = df_all['ocupacaoConfirmadoCli'].astype('int')
df_all['ocupacaoConfirmadoUti'] = df_all['ocupacaoConfirmadoUti'].astype('int')
df_all['ocupacaoCovidUti'] = df_all['ocupacaoCovidUti'].astype('int')
df_all['ocupacaoCovidCli'] = df_all['ocupacaoCovidCli'].astype('int')
df_all['ocupacaoHospitalarUti'] = df_all['ocupacaoHospitalarUti'].astype('int')
df_all['ocupacaoHospitalarCli'] = df_all['ocupacaoHospitalarCli'].astype('int')
df_all['saidaSuspeitaObitos'] = df_all['saidaSuspeitaObitos'].astype('int')
df_all['saidaSuspeitaAltas'] = df_all['saidaSuspeitaAltas'].astype('int')
df_all['saidaConfirmadaObitos'] = df_all['saidaConfirmadaObitos'].astype('int')
df_all['ocupacaoSuspeitoCli '] = df_all['ocupacaoSuspeitoCli'].astype('int')
```

```python
df_all = df_all.groupby(['datacode']).sum()
df_all
```

```python
df_all.reset_index(inplace=True)
```

```python
df_all.to_csv('brasil_ocupacao_daily.csv')
```
